```python
import pandas as pd
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
# here we are applying ordinal encoding on ordinal, gender and label encoding on purchase
df = df.iloc[:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size= 0.2)
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_test= oe.transform(X_test)
X_train = oe.transform(X_train) # this will result in array hence we have to convert to dataFrame
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # here you do not get to choose which value to assign to what
le.fit(y_train)
le.classes_
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('titanic.csv')
df
```

##### 1. SCATTER PLOT (NUMERICAL-NUMERICAL)
###### PAIR PLOT(VERY USEFUL), LINE PLOT CAN ALSO BE USED 

```python
sns.scatterplot(x=df['Age'], y=df['Fare'], hue=df['Sex'], style=df['Survived'], size=df['Embarked']) # more than 2 column analysis- multivariate
```

##### BAR PLOT (NUMERICAL-CATEGORICAL)

```python
sns.barplot(x=df['Pclass'], y=df['Fare'],hue=df['Sex'] )
```

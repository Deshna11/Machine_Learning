##### ONE_HOT_ENCODING WITH TOP CATEGORIES

```python
counts = df['brand'].value_counts()
threshold = 100
repl = counts[counts <= threshold].index
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype=np.int32)
```
